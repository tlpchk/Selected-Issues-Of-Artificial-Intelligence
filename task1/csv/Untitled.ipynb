```python
import pandas as pd
```

```python
df = pd.read_csv("./coin_mcts.csv")
df.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3))

xlabel = "n"
ylabel = "avg. convincity"
title = "Coin MCTS"

x = df[xlabel]
y = df[ylabel]

plt.bar(x,y,align='center')

plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.title(title)
plt.show()
```

```python
import seaborn as sns
N = 10
x = list(range(1,N+1))

p_10 = list(pd.read_csv("10_ms.csv")["avg. playouts"])
p_30 = list(pd.read_csv("30_ms.csv")["avg. playouts"])
p_100 = list(pd.read_csv("100_ms.csv")["avg. playouts"])

df = pd.DataFrame(zip(x*N, ["10"]*N+["30"]*N+["100"]*N, p_10+p_30+p_100), columns=["n", "time (ms)", "avg. plyouts"])
plt.figure(figsize=(10, 6))
sns.barplot(x="n", hue="time (ms)", y="avg. plyouts", data=df)
plt.show()
```
