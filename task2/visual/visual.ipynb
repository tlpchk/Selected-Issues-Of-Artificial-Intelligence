```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
N = 15
x = list(range(1,N+1))

df_10 = pd.read_csv("q_mcts_vs_mcts_10.csv",index_col='n')
df_50 = pd.read_csv("q_mcts_vs_mcts_50.csv",index_col='n')
df_full = pd.concat([df_10, df_50], axis=1)

mcts_10 = df_full['mcts_10'].to_list()[:N]
qmcts_10 = df_full['q_mcts_10'].to_list()[:N]
mcts_50 = df_full['mcts_50'].to_list()[:N]
qmcts_50 = df_full['q_mcts_50'].to_list()[:N]

df = pd.DataFrame(zip(x*N, ["MCTS 10ms"]*N+["Q-MCTS 10ms"]*N , mcts_10+qmcts_10), columns=["n", "time", "avg. plyouts"])
plt.figure(figsize=(10, 6))
sns.barplot(x="n", hue="time (ms)", y="avg. plyouts", data=df)
plt.show()

df = pd.DataFrame(zip(x*N, ["MCTS 50ms"]*N + ["Q-MCTS 50ms"]*N , mcts_50+qmcts_50), columns=["n", "time", "avg. plyouts"])
plt.figure(figsize=(10, 6))
sns.barplot(x="n", hue="time (ms)", y="avg. plyouts", data=df)
plt.show()

```
